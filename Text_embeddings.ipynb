```python
from sentence_transformers import SentenceTransformer

# Initialize the model
model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')

# Your text
our_text = "Hey buddy"

# Generate embeddings
text_embedding = model.encode(our_text)

# Print the embedding
print(f"Our embedding is {text_embedding}")
```

```python
import pandas as pd
data = ['School', 'college', 'car', 'bike', 'apple', 'orange','banana']
df = pd.DataFrame(data, columns=['words'])
df
```

```python
df['embeddings'] = df['words'].apply(lambda x: model.encode(x))
df.to_csv('word_embeddings.csv')
```

```python
df
```

```python
our_text = 'mango'

# Generate embeddings
text_embedding = model.encode(our_text).reshape(1, -1)

# Print the embedding
print(f"Our embedding is {text_embedding}")
```

```python
from sklearn.metrics.pairwise import cosine_similarity

df['similarity_score'] = df['embeddings'].apply(lambda x: cosine_similarity(x.reshape(1, -1), text_embedding).flatten()[0])
```
